```python
# Import the necessary libraries
import os
os.chdir("..")
from config import cfg
import random
```

```python
ecoli_input_file = cfg["data"]["fragmented_ecoli"]
yeast_input_file = cfg["data"]["fragmented_yeast"]
mixture_input_file = cfg["data"]["fragmented_mixture"]
ecoli_train_split = cfg["data"]["ecoli_train_split"]
yeast_train_split = cfg["data"]["yeast_train_split"]
mixture_train_split = cfg["data"]["mixture_train_split"]
ecoli_test_split = cfg["data"]["ecoli_test_split"]
yeast_test_split = cfg["data"]["yeast_test_split"]
mixture_test_split = cfg["data"]["mixture_test_split"]
test_ratio = cfg["data"]["test_ratio"]
random.seed(cfg["misc"]["seed"])
```

```python
# Method to shuffle and split the data into training and testing sets
def split_data(input_file, train_split, test_split, test_ratio):
    with open(input_file, "r") as f:
        lines = f.readlines()
    
    random.shuffle(lines)
    
    split_index = int(len(lines) * (1 - test_ratio))
    train_lines = lines[:split_index]
    test_lines = lines[split_index:]
    
    with open(train_split, "w") as f:
        f.writelines(train_lines)
    
    with open(test_split, "w") as f:
        f.writelines(test_lines)
```

```python
# Split the E. coli and yeast datasets
split_data(ecoli_input_file, ecoli_train_split, ecoli_test_split, test_ratio)
split_data(yeast_input_file, yeast_train_split, yeast_test_split, test_ratio)
split_data(mixture_input_file, mixture_train_split, mixture_test_split, test_ratio)
```
