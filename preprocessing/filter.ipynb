```python
# Import the necessary libraries
from Bio import SeqIO
from collections import Counter
```

```python
data_dir = "../data/uniprot_sprot.fasta"  # path to the input FASTA file
output_dir = "../data/ecoli_yeast.fasta"  # path to the output FASTA file
patterns = ["Escherichia coli", "Saccharomyces cerevisiae"]  # patterns to search for
record_iterator = SeqIO.parse(data_dir, "fasta")
counter = Counter()

with open(output_dir, "w") as output_handle:
    for record in record_iterator:
        for pattern in patterns:
            if pattern in record.description:
                counter[pattern] += 1
                SeqIO.write(record, output_handle, "fasta")
                break

for pattern, count in counter.items():
    print(f"{pattern}: {count} records")
```
